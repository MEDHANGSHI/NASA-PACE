```python
!pip install transformers torch

from transformers import BertForQuestionAnswering, BertTokenizer
import torch

# Load the pre-trained model and tokenizer
model = BertForQuestionAnswering.from_pretrained('bert-large-uncased-whole-word-masking-finetuned-squad')
tokenizer = BertTokenizer.from_pretrained('bert-large-uncased-whole-word-masking-finetuned-squad')

def answer_question(question, context):
    inputs = tokenizer(question, context, return_tensors='pt')
    with torch.no_grad():
        outputs = model(**inputs)

    # Get the most likely start and end of the answer
    answer_start = torch.argmax(outputs.start_logits)
    answer_end = torch.argmax(outputs.end_logits) + 1

    # Convert tokens to string
    answer = tokenizer.convert_tokens_to_string(tokenizer.convert_ids_to_tokens(inputs.input_ids[0][answer_start:answer_end]))
    return answer

def answer_questions(questions, context):
    answers = {}
    for question in questions:
        answer = answer_question(question, context)
        answers[question] = answer
    return answers

# Original context and questions
original_context = """
NASA’s PACE (Plankton, Aerosol, Cloud, ocean Ecosystem) mission is designed to advance our understanding of Earth’s ocean-atmosphere system.
It will observe ocean ecosystems, airborne particles, and clouds, and provide insights into the carbon cycle and climate change.
"""

original_questions = [
    "What does PACE mission aim to observe?",
    "What are the main components of the PACE mission?",
    "How does the PACE mission contribute to understanding climate change?"
]

# Get answers for the original context and questions
original_answers = answer_questions(original_questions, original_context)
print("Original Context Answers:")
for question, answer in original_answers.items():
    print(f"Question: {question}\nAnswer: {answer}\n")

# New context and questions
new_context = """
Clouds play a critical role in Earth's climate system. They help regulate temperature by reflecting sunlight and trapping heat.
Clouds come in various types including cumulus, stratus, and cirrus, each with distinct characteristics.
Understanding cloud formation and behavior is essential for accurate climate models and weather predictions.
"""

new_questions = [
    "What are the different types of clouds mentioned?",
    "Why is understanding cloud formation important?"
]

# Get answers for the new context and questions
new_answers = answer_questions(new_questions, new_context)
print("New Context Answers:")
for question, answer in new_answers.items():
    print(f"Question: {question}\nAnswer: {answer}\n")
```
